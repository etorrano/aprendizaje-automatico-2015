5 Un modo de utilizar los datos de texto en aprendizaje es pasar a un modelo vectorial. Explique brevemente como es el modelo vectorial con *bag-of-words*, qué hace *CountVectorizer* definido en sklearn, qué es *tf-idf* y cómo se implementa en Python. 

El modelo vectorial con *bag-of-words*, le asigna a cada palabra un identificador y guarda en una matriz, que es esparsa, la cantidad de ocurrencia de la palabra en cada documento. En la posición i,j se guarda el número de ocurrencias con i,j los identificadores del documento y la palabra respectivamente.
*CountVectorizer* genera un diccionario de N-gramas o por caracter permitiendo tenerlos indizados para el procesamiento. El método *tf-idf* consiste en asociar una frecuencia a cada palabra dividido por la cantidad de palabras en el documento. Además se le da menos importancia a las palabras que aparecen más veces en el cuerpo que las demás porque se considera que aportan menos información para realizar la clasificación del documento.
Para implementar éste último se utiliza una función *fit_transform(..)* la cuál ajusta el estimador del método a los datos y transforma la matriz mencionada anteriormente en otra representación más adecuada al método.

```python
import numpy as np
#convierto vocabulario en arreglo
voc_arr = vec_vocab.toarray()
#sumo cada fila para obtener la cantidad de palabras por documento
dist = np.sum(voc_arr, axis=1)
#imprimo sumatoria por cada documento
print dist
#máxima cantidad de palabras
print "El largo máximo de un documento es:" + str(max(dist))
```
